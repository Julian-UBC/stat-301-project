## Data:

We decided to use this dataset for our project: https://archive.ics.uci.edu/dataset/275/bike+sharing+dataset

This dataset contains the hourly and daily count of rental bikes between years 2011 and 2012 in Capital bikeshare system with the corresponding weather and seasonal information.

Fanaee-T (2013) collected the data from Capital bikeshare system in Washington D.C. during the 2011-2012 year period.

```r
temp <- tempfile()
download.file("https://archive.ics.uci.edu/static/public/275/bike+sharing+dataset.zip", temp)
data_day <- read.csv(unz(temp, "day.csv"), header = TRUE, sep = ",") 
data_hour <- read.csv(unz(temp, "hour.csv"), header = TRUE, sep = ",") 
unlink(temp)
```

`data_day` contains the daily count data. Below we can see the first 6 rows of `data_day`.

```r
head(data_day)
```

`data_hour` contains the hourly count data. Below we can see the first 6 rows of`data_hour`.

```r
head(data_hour)
```

```r
nrow(data_hour)
nrow(data_day)
```

`data_hour` has a total of 17379 observations with no missing values. `data_day` has a total of 731 observations with no missing values.

`data_hour` has 17 variables while `data_day` has 16 variables. Both `data_hour` and `data_day` have the same following variables, except `hr` which is only in `data_hour`.

- instant (int): record index
- dteday (chr): date
- season (int): season (1:winter, 2:spring, 3:summer, 4:fall)
- yr (int): year (0: 2011, 1:2012)
- mnth (int): month (1 to 12)
- hr (int): hour (0 to 23) [only for `data_hour`]
- holiday (int): whether day is holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
- weekday (int): day of the week
- workingday (int): if day is neither weekend nor holiday is 1, otherwise is 0.
- weathersit (int): 
    - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp (dbl): Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)
- atemp (dbl): Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)
- hum (dbl): Normalized humidity. The values are divided to 100 (max)
- windspeed (dbl): Normalized wind speed. The values are divided to 67 (max)
- casual (int): count of casual users
- registered (int): count of registered users
- cnt (int): count of total rental bikes including both casual and registered



## Question:

What is the expected total count of rental bikes daily based on season, working day, and temperature?

The data could help us answer this question because they have the variables needed to answer our question in each observation. In this case, our response variable is `cnt` and our explanatory variables are `season`, `temp`, and `workingday`. 

## Data Cleaning and Wrangling

Since we want to predict the daily count of rental bikes, we are going to use `data_day`. We also want to only select the columns needed for our prediction (`cnt`, `season`, `temp`, and `workingday`). 

Since both `season` and `workingday` are categorical variables (rather than continuous variables), it would be better if we change the type from `int` to `fct`.

The (final) dataset will be stored in `bike_data`.

```r
bike_data <- data_day %>% select(cnt, season, temp, workingday)
bike_data$season <- as.factor(bike_data$season)
bike_data$workingday <- as.factor(bike_data$workingday)

head(bike_data)
```

## Visualization:

Before we dive deeper, we want to first visualize the relationship between each variables in `bike_data`. We could use this information to help decide the model that we would want to use later. To achieve this, we would use `ggpairs()` to generate a pair plot of ALL the variables found in `bike_data`.

```r
options(repr.plot.width = 13, repr.plot.height = 12) # Adjust these numbers so the plot looks good in your desktop.

bike_pair_plots <- bike_data %>%
  ggpairs(progress = FALSE) +
  theme(
    text = element_text(size = 15),
    plot.title = element_text(face = "bold"),
    axis.title = element_text(face = "bold")
  )

bike_pair_plots
```

From the plots above, we can see the relationship between each variables, especially how the response variable (`cnt`) relates to each of the input variables (`season`, `workingday`, `temp`).

- There is a positive linear relationship between `cnt` and `temp` with a correlation of approximately 0.63.

- The boxplots above for `cnt` and `season` show that some seasons have some differences in daily count of rental bikes.

- The boxplots above for `cnt` and `workingday` show no differences in daily count of rental bikes between working and non-working days.

##  Methods and Plan

I have decided to use multiple linear regression (MLR) with an additive model to address my question of interest. I will be using the regression formula `cnt ~.` to encompass all available variables for analysis. While my initial focus is on understanding the predictive power of `season`, `workingday`, and `temp` for `cnt`, I am equally interested in identifying any variables that demonstrate statistically significant associations with the response variable.

MLR allows the creation of a robust model for predicting `cnt` based on specified input values. However, the application of MLR requires adherence to five assumptions. These include the assumption of a linear relationship between `cnt` and the input variables, independence of errors, a normal conditional distribution of errors, equal variance of errors, and independence among input variables to prevent multicollinearity.

While these assumptions are crucial for the effectiveness of the MLR model, they also come with potential weaknesses. Specifically, linear regression assumes a linear relationship between the response and input variables, which may result in poor fits when the actual relationship is nonlinear. Additionally, the vulnerability of linear regression to outliers is a concern, especially in datasets with limited data points. Outliers can disproportionately affect estimated coefficients, influencing the overall interpretability of the model. It's important to acknowledge and address these limitations for a more nuanced and well-informed interpretation of the results.
